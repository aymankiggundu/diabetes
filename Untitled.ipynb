```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Suppress FutureWarnings and other potential warnings
warnings.filterwarnings("ignore", category=FutureWarning)

# Load only the first sub-dataset
file_path = "aymankiggundu/Desktop/BCSC/year_two/Year two Semester 2 2025/Machine Learning/Assignments/diabetes/sub_dataset_1.csv"
data = pd.read_csv(file_path)

# Display the first few rows
print("Dataset Preview:")
print(data.head())

# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())

# Summary statistics
print("\nSummary Statistics:")
print(data.describe())
```

```python
# Scatter plot of Glucose Level vs. EGG Frequency
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['Glucose_Level'], y=data['EGG_Frequency'], hue=data['Diabetes_Status'], palette={0: 'green', 1: 'red'})
plt.title("Glucose Level vs. EGG Frequency")
plt.xlabel("Glucose Level (mg/dL)")
plt.ylabel("EGG Frequency (Hz)")
legend_labels = {0: "Healthy", 1: "Diabetic"}
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, [legend_labels[int(label)] for label in labels], title="Diabetes Status")
plt.show()

# Box plot of Glucose Level by Diabetes Status
plt.figure(figsize=(10, 6))
sns.boxplot(x='Diabetes_Status', y='Glucose_Level', data=data, hue='Diabetes_Status', palette=['green', 'red'])
plt.title("Glucose Level Distribution by Diabetes Status")
plt.xlabel("Diabetes Status (0 = No, 1 = Yes)")
plt.ylabel("Glucose Level (mg/dL)")
plt.xticks(ticks=[0, 1], labels=["Healthy", "Diabetic"])
plt.show()

# Line plot for EGG Amplitude variation per Patient_ID
plt.figure(figsize=(10, 6))
sns.lineplot(x=data['Patient_ID'], y=data['EGG_Amplitude'], marker='o', linestyle='-', color='blue')
plt.title("EGG Amplitude Variation Across Patients")
plt.xlabel("Patient ID")
plt.ylabel("EGG Amplitude (mV)")
plt.show()

# Diabetes prevalence
plt.figure(figsize=(10, 6))
sns.countplot(x='Diabetes_Status', data=data, hue='Diabetes_Status', palette=['green', 'red'])
plt.title("Diabetes Prevalence in Dataset")
plt.xlabel("Diabetes Status")
plt.ylabel("Count")
plt.xticks(ticks=[0, 1], labels=["Healthy", "Diabetic"])
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()

# 2. Correlation Heatmap with Target Variable (Diabetes_Status)
correlation = data.corr()
target_corr = correlation['Diabetes_Status'].sort_values(ascending=False)

plt.figure(figsize=(6, 4))
sns.barplot(x=target_corr.index, y=target_corr.values, palette='coolwarm')
plt.title("Correlation of Features with Diabetes Status")
plt.xlabel("Features")
plt.ylabel("Correlation Coefficient")
plt.xticks(rotation=45)
plt.show()

# 3. Violin Plot of Glucose Level by Diabetes Status
plt.figure(figsize=(8, 6))
sns.violinplot(x='Diabetes_Status', y='Glucose_Level', data=data, palette=['green', 'red'])
plt.title("Distribution of Glucose Levels by Diabetes Status")
plt.xlabel("Diabetes Status")
plt.ylabel("Glucose Level (mg/dL)")
plt.xticks(ticks=[0, 1], labels=["Healthy", "Diabetic"])
plt.show()

# 4. Scatter Plot of EGG Frequency vs. EGG Amplitude
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['EGG_Frequency'], y=data['EGG_Amplitude'], hue=data['Diabetes_Status'], palette={0: 'green', 1: 'red'})
plt.title("EGG Frequency vs. EGG Amplitude")
plt.xlabel("EGG Frequency (Hz)")
plt.ylabel("EGG Amplitude (mV)")
plt.legend(title="Diabetes Status", labels=["Healthy", "Diabetic"])
plt.show()

# 5. Heatmap for Feature Correlation (only relevant features)
relevant_features = ['Glucose_Level', 'EGG_Frequency', 'EGG_Amplitude', 'Diabetes_Status']
plt.figure(figsize=(8, 6))
sns.heatmap(data[relevant_features].corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Heatmap (Relevant Features)")
plt.show()

 
```
